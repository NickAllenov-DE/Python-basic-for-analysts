#### Задание 2

Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions

```python
# Самый простой способ, но без удаления элементов

add_costs = [100, 125, -90, 345, 655, -1, 0, 200]
print(f'Общие зартраты на рекламу - {sum(i for i in add_costs if i > 0)}')
```

```python
# То же простой способ, но опять же без удаления элементов, а с помощью создания нового списка

add_costs = [100, 125, -90, 345, 655, -1, 0, 200]
corr_costs = [i for i in add_costs if i > 0]
print(f'corrected costs - {corr_costs}')
print(f'Total cost = {sum(corr_costs)}')
```

```python
# Если нужно удалить неправильные элементы из оригинального списка, то коротко не получается (+ много лишних итераций, которые 
# могут значительно замедлить расчёты при большом объеме данных, но это необходимо в связи со смещением индексов элементов при
# удалении их из оригинального списка):

add_costs = [100, 125, -90, 345, 655, -1, 0, 200]
for i in add_costs:
    for i in add_costs:
        if i <= 0:
            add_costs.remove(i)
print(add_costs)
print(f'Total cost = {sum(add_costs)}')
```
